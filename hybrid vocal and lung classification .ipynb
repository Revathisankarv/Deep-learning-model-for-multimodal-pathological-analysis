```python
import os
import numpy as np
import librosa
import librosa.display
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, LSTM, Bidirectional, Dropout, Conv1D, MaxPooling1D, GRU, Input, concatenate
from tensorflow.keras.optimizers import RMSprop, Adam
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.layers import BatchNormalization
import random
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.regularizers import l1_l2, l2

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)
random.seed(42)

# Data Augmentation Functions
def add_noise(data, noise_factor=0.005):
    noise = np.random.randn(len(data))
    augmented_data = data + noise_factor * noise
    return augmented_data

def time_shift(data, shift_max=0.2):
    shift = np.random.randint(len(data) * shift_max)
    return np.roll(data, shift)

def change_pitch(data, sr, n_steps=2):
    return librosa.effects.pitch_shift(y=data, sr=sr, n_steps=n_steps)

def stretch(data, rate=0.8):
    return librosa.effects.time_stretch(y=data, rate=rate)

def random_erase(data, erase_percentage=0.1):
    data_length = len(data)
    erase_length = int(data_length * erase_percentage)
    erase_start = np.random.randint(0, data_length - erase_length)
    data[erase_start:erase_start + erase_length] = 0
    return data

def speed_tuning(data, rate=2):
    return librosa.effects.time_stretch(data, rate=rate)

def mixup(data1, data2, alpha=0.2):
    lam = np.random.beta(alpha, alpha)
    min_length = min(len(data1), len(data2))
    data1 = data1[:min_length]
    data2 = data2[:min_length]
    mixed_data = lam * data1 + (1 - lam) * data2
    return mixed_data, lam

# Feature Extraction Function
def extract_features(file_path, sr=22050, n_mfcc=13):
    audio, _ = librosa.load(file_path, sr=sr)
    
    # Original
    mfccs = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=n_mfcc)
    feature = np.mean(mfccs.T, axis=0)
    
    return feature, audio, sr

# Modified Load Data Function to create balanced datasets
def load_balanced_data(normal_path, pathological_path, target_count=6448, is_lung=False):
    normal_features = []
    pathological_features = []
    
    normal_audios = []
    pathological_audios = []
    
    normal_srs = []
    pathological_srs = []
    
    # Load original data
    print(f"Loading {'lung' if is_lung else 'voice'} data...")
    
    # Load normal samples
    for file in os.listdir(normal_path):
        if file.endswith('.wav'):
            file_path = os.path.join(normal_path, file)
            feature, audio, sr = extract_features(file_path)
            normal_features.append(feature)
            normal_audios.append(audio)
            normal_srs.append(sr)
    
    # Load pathological samples
    for file in os.listdir(pathological_path):
        if file.endswith('.wav'):
            file_path = os.path.join(pathological_path, file)
            feature, audio, sr = extract_features(file_path)
            pathological_features.append(feature)
            pathological_audios.append(audio)
            pathological_srs.append(sr)
    
    original_normal_count = len(normal_features)
    original_pathological_count = len(pathological_features)
    
    print(f"Original {'lung' if is_lung else 'voice'} data: Normal: {original_normal_count}, Pathological: {original_pathological_count}")
    
    # Augmentation functions list with varying parameters
    def get_augmentation_functions():
        return [
            lambda x, sr: (add_noise(x, noise_factor=np.random.uniform(0.001, 0.01)), sr),  
            lambda x, sr: (time_shift(x, shift_max=np.random.uniform(0.1, 0.3)), sr),  
            lambda x, sr: (change_pitch(x, sr, n_steps=np.random.uniform(-3, 3)), sr),  
            lambda x, sr: (stretch(x, rate=np.random.uniform(0.8, 1.2)), sr),  
            lambda x, sr: (random_erase(x.copy(), erase_percentage=np.random.uniform(0.05, 0.15)), sr),  
            lambda x, sr: (speed_tuning(x, rate=np.random.uniform(0.85, 1.15)), sr),  
        ]

    # Function to generate multiple augmentations of a single audio
    def generate_augmentations(audio_list, sr_list, features_list, target_count):
        current_count = len(features_list)
        
        if current_count >= target_count:
            indices = random.sample(range(current_count), target_count)
            return [features_list[i] for i in indices]
        
        # Calculate how many augmentations 
        augmentations_needed = target_count - current_count
        augmentations_per_sample = max(1, int(np.ceil(augmentations_needed / current_count)))
        
        aug_functions = get_augmentation_functions()
        new_features = features_list.copy()
        
        while len(new_features) < target_count:
            idx = random.randrange(len(audio_list))
            audio = audio_list[idx]
            sr = sr_list[idx]
            
            # Choose a random augmentation function
            aug_func = random.choice(aug_functions)
            aug_audio, aug_sr = aug_func(audio, sr)
            
            # Extract features from augmented audio
            mfccs = librosa.feature.mfcc(y=aug_audio, sr=aug_sr, n_mfcc=13)
            feature = np.mean(mfccs.T, axis=0)
            
            new_features.append(feature)
            
            # For mixup augmentation (reduced probability to 0.15)
            if random.random() < 0.15 and len(audio_list) > 1:
                idx2 = random.randrange(len(audio_list))
                while idx2 == idx:  
                    idx2 = random.randrange(len(audio_list))
                
                audio2 = audio_list[idx2]
                min_length = min(len(audio), len(audio2))
                audio = audio[:min_length]
                audio2 = audio2[:min_length]
                
                mixed_audio, _ = mixup(audio, audio2, alpha=0.2)
                mfccs = librosa.feature.mfcc(y=mixed_audio, sr=sr, n_mfcc=13)
                feature = np.mean(mfccs.T, axis=0)
                new_features.append(feature)
            
            if len(new_features) >= target_count:
                break
        
        # Ensure exactly target_count features
        return new_features[:target_count]
    
    # Generate balanced datasets
    normal_features_balanced = generate_augmentations(normal_audios, normal_srs, normal_features, target_count)
    pathological_features_balanced = generate_augmentations(pathological_audios, pathological_srs, pathological_features, target_count)
    
    # Combine features and labels
    features = normal_features_balanced + pathological_features_balanced
    labels = ['normal'] * target_count + ['pathological'] * target_count
    
    print(f"Balanced {'lung' if is_lung else 'voice'} dataset: {target_count} normal samples, {target_count} pathological samples")
    print(f"Total {'lung' if is_lung else 'voice'} samples: {len(features)}")
    
    return np.array(features), np.array(labels)

# Define Paths
voice_normal_path = 'database/normal'
voice_pathological_path = 'database/abnormal'
lung_normal_path = 'database/respiratory data set/Healthy'
lung_pathological_path = 'database/respiratory data set/abnormal'

# Target count for balanced datasets 
TARGET_COUNT = 4000  
TOTAL_TARGET = TARGET_COUNT * 2  

# Load Balanced Data
X_voice, y_voice = load_balanced_data(voice_normal_path, voice_pathological_path, TARGET_COUNT)
X_lung, y_lung = load_balanced_data(lung_normal_path, lung_pathological_path, TARGET_COUNT, is_lung=True)

# Print total dataset information
print("\nDataset Summary:")
print(f"Voice dataset: {len(X_voice)} total samples ({y_voice.tolist().count('normal')} normal, {y_voice.tolist().count('pathological')} pathological)")
print(f"Lung dataset: {len(X_lung)} total samples ({y_lung.tolist().count('normal')} normal, {y_lung.tolist().count('pathological')} pathological)")
print(f"Combined dataset: {len(X_voice) + len(X_lung)} total samples")

# Encode the Labels
label_encoder = LabelEncoder()
y_voice_encoded = label_encoder.fit_transform(y_voice)
y_lung_encoded = label_encoder.fit_transform(y_lung)

# Ensure same random state for consistent splitting
random_state = 42
X_voice_train, X_voice_temp, y_voice_train, y_voice_temp = train_test_split(
    X_voice, y_voice_encoded, test_size=0.3, random_state=random_state, stratify=y_voice_encoded
)
# Split temp into validation and test sets
X_voice_val, X_voice_test, y_voice_val, y_voice_test = train_test_split(
    X_voice_temp, y_voice_temp, test_size=0.5, random_state=random_state, stratify=y_voice_temp
)

# Do the same for lung data
X_lung_train, X_lung_temp, y_lung_train, y_lung_temp = train_test_split(
    X_lung, y_lung_encoded, test_size=0.3, random_state=random_state, stratify=y_lung_encoded
)
X_lung_val, X_lung_test, y_lung_val, y_lung_test = train_test_split(
    X_lung_temp, y_lung_temp, test_size=0.5, random_state=random_state, stratify=y_lung_temp
)

# Print split information
print("\nSplit Summary:")
print(f"Voice training: {len(X_voice_train)} samples")
print(f"Voice validation: {len(X_voice_val)} samples")
print(f"Voice testing: {len(X_voice_test)} samples")
print(f"Lung training: {len(X_lung_train)} samples")
print(f"Lung validation: {len(X_lung_val)} samples")
print(f"Lung testing: {len(X_lung_test)} samples")

# Reshape Features for LSTM Input
X_voice_train = X_voice_train[..., np.newaxis]
X_voice_val = X_voice_val[..., np.newaxis]
X_voice_test = X_voice_test[..., np.newaxis]
X_lung_train = X_lung_train[..., np.newaxis]
X_lung_val = X_lung_val[..., np.newaxis]
X_lung_test = X_lung_test[..., np.newaxis]

# Set up callbacks to prevent overfitting
callbacks = [
    EarlyStopping(
        monitor='val_loss',
        patience=10,
        restore_best_weights=True,
        verbose=1
    ),
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=5,
        min_lr=1e-6,
        verbose=1
    ),
    ModelCheckpoint(
        filepath='best_model_{epoch:02d}_{val_loss:.4f}.h5',
        monitor='val_loss',
        save_best_only=True,
        verbose=1
    )
]

# Build RNN-BiLSTM Model for Voice Data with L1L2 regularization
reg = l2(l2=1e-4)  
voice_input = Input(shape=(X_voice_train.shape[1], 1))
voice_x = Conv1D(32, kernel_size=3, activation='relu', kernel_regularizer=reg)(voice_input)
voice_x = BatchNormalization()(voice_x) 
voice_x = MaxPooling1D(pool_size=2)(voice_x)
voice_x = Bidirectional(LSTM(64, return_sequences=True, kernel_regularizer=reg))(voice_x)
voice_x = BatchNormalization()(voice_x) 
voice_x = Dropout(0.3)(voice_x)  
voice_x = Bidirectional(GRU(32, kernel_regularizer=reg))(voice_x) 
voice_x = BatchNormalization()(voice_x) 
voice_x = Dropout(0.3)(voice_x)
voice_output = Dense(1, activation='sigmoid', kernel_regularizer=reg)(voice_x)

voice_model = Model(inputs=voice_input, outputs=voice_output)
voice_model.compile(optimizer=RMSprop(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

voice_model.summary()

# Train the Voice Model with validation data and callbacks
voice_history = voice_model.fit(
    X_voice_train, y_voice_train,
    epochs=100, 
    batch_size=16, 
    validation_data=(X_voice_val, y_voice_val),
    callbacks=callbacks
)

# Build RNN-BiLSTM Model for Lung Data 
lung_input = Input(shape=(X_lung_train.shape[1], 1))
lung_x = Conv1D(32, kernel_size=3, activation='relu', kernel_regularizer=reg)(lung_input)
lung_x = BatchNormalization()(lung_x) 
lung_x = MaxPooling1D(pool_size=2)(lung_x)
lung_x = Bidirectional(LSTM(64, return_sequences=True, kernel_regularizer=reg))(lung_x)
lung_x = BatchNormalization()(lung_x) 
lung_x = Dropout(0.3)(lung_x)
lung_x = Bidirectional(GRU(32, kernel_regularizer=reg))(lung_x)
lung_x = BatchNormalization()(lung_x) 
lung_x = Dropout(0.3)(lung_x)
lung_output = Dense(1, activation='sigmoid', kernel_regularizer=reg)(lung_x)

lung_model = Model(inputs=lung_input, outputs=lung_output)
lung_model.compile(optimizer=RMSprop(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

lung_model.summary()

# Train the Lung Model with validation data and callbacks
lung_history = lung_model.fit(
    X_lung_train, y_lung_train,
    epochs=100,
    batch_size=16,
    validation_data=(X_lung_val, y_lung_val),
    callbacks=callbacks
)

# Predict with both models for training, validation and test sets
voice_pred_train = voice_model.predict(X_voice_train).flatten()
voice_pred_val = voice_model.predict(X_voice_val).flatten()
voice_pred_test = voice_model.predict(X_voice_test).flatten()

lung_pred_train = lung_model.predict(X_lung_train).flatten()
lung_pred_val = lung_model.predict(X_lung_val).flatten()
lung_pred_test = lung_model.predict(X_lung_test).flatten()

# Ensure the lengths match for training
min_train_len = min(len(voice_pred_train), len(lung_pred_train))
combined_train = np.vstack((voice_pred_train[:min_train_len], lung_pred_train[:min_train_len])).T
y_combined_train = y_voice_train[:min_train_len]

# Ensure the lengths match for validation
min_val_len = min(len(voice_pred_val), len(lung_pred_val))
combined_val = np.vstack((voice_pred_val[:min_val_len], lung_pred_val[:min_val_len])).T
y_combined_val = y_voice_val[:min_val_len]

# Ensure the lengths match for testing
min_test_len = min(len(voice_pred_test), len(lung_pred_test))
combined_test = np.vstack((voice_pred_test[:min_test_len], lung_pred_test[:min_test_len])).T
y_combined_test = y_voice_test[:min_test_len]

# Build Ensemble Model with regularization
final_input = Input(shape=(2,))
final_x = Dense(32, activation='relu', kernel_regularizer=reg)(final_input) 
final_x = BatchNormalization()(final_x)
final_x = Dropout(0.4)(final_x)  
final_output = Dense(1, activation='sigmoid', kernel_regularizer=reg)(final_x)

final_model = Model(inputs=final_input, outputs=final_output)
final_model.compile(optimizer=Adam(learning_rate=0.0005), loss='binary_crossentropy', metrics=['accuracy'])

final_model.summary()

# Train the Ensemble Model with validation data and callbacks
final_history = final_model.fit(
    combined_train, y_combined_train,
    epochs=50,
    batch_size=16,
    validation_data=(combined_val, y_combined_val),
    callbacks=callbacks
)

# Function to plot training history
def plot_training_history(history, title):
    plt.figure(figsize=(12, 5))
    
    # Plot training & validation accuracy values
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title(f'{title} Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='lower right')
    
    # Plot training & validation loss values
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title(f'{title} Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper right')
    
    plt.tight_layout()
    plt.show()

# Function to evaluate model
def evaluate_model(model, X_test, y_test, label_encoder, title):
    y_pred = model.predict(X_test)
    y_pred_classes = np.round(y_pred).astype(int).reshape(-1)
    print(f'Classification Report for {title} Data:')
    print(classification_report(y_test, y_pred_classes, target_names=label_encoder.classes_))
    
    # Compute and display confusion matrix
    cm = confusion_matrix(y_test, y_pred_classes)
    plt.figure(figsize=(10,7))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=label_encoder.classes_, 
                yticklabels=label_encoder.classes_)
    plt.title(f'Confusion Matrix for {title} Data')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()
    
    # Display normalized confusion matrix
    cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    plt.figure(figsize=(10,7))
    sns.heatmap(cm_normalized, annot=True, fmt='.2f', cmap='Blues', 
                xticklabels=label_encoder.classes_, 
                yticklabels=label_encoder.classes_)
    plt.title(f'Normalized Confusion Matrix for {title} Data')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

# Function to plot ROC curve
def plot_roc_curve(y_true, y_pred, title):
    fpr, tpr, _ = roc_curve(y_true, y_pred)
    roc_auc = auc(fpr, tpr)
    
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (AUC = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'Receiver Operating Characteristic (ROC) Curve - {title}')
    plt.legend(loc="lower right")
    plt.show()

# Plot training and validation graphs
print("\n===== TRAINING HISTORY PLOTS =====")
plot_training_history(voice_history, 'Voice')
plot_training_history(lung_history, 'Lung')
plot_training_history(final_history, 'Ensemble')

# Evaluate Voice Model
print("\n===== VOICE MODEL EVALUATION =====")
evaluate_model(voice_model, X_voice_test, y_voice_test, label_encoder, 'Voice')
voice_pred_proba = voice_model.predict(X_voice_test).flatten()
plot_roc_curve(y_voice_test, voice_pred_proba, 'Voice Model')

# Evaluate Lung Model
print("\n===== LUNG MODEL EVALUATION =====")
evaluate_model(lung_model, X_lung_test, y_lung_test, label_encoder, 'Lung')
lung_pred_proba = lung_model.predict(X_lung_test).flatten()
plot_roc_curve(y_lung_test, lung_pred_proba, 'Lung Model')

# Evaluate the Ensemble Model
print("\n===== ENSEMBLE MODEL EVALUATION =====")
evaluate_model(final_model, combined_test, y_combined_test, label_encoder, 'Combined Voice and Lung')
ensemble_pred_proba = final_model.predict(combined_test).flatten()
plot_roc_curve(y_combined_test, ensemble_pred_proba, 'Combined Voice and Lung Model')
```
